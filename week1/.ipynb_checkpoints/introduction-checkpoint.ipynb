#### 强人工智能(BOTTOM-UP AI)
强人工智能观点认为有可能制造出真正能推理（REASONING）和解决问题（PROBLEM_SOLVING）的智能机器，<br>
并且，这样的机器能将被认为是有知觉的，有自我意识的。强人工智能可以有两类：
`类人的人工智能`，即机器的思考和推理就像人的思维一样。<br>
`非类人的人工智能`，即机器产生了和人完全不一样的知觉和意识，使用和人完全不一样的推理方式。<br>
#### 弱人工智能(TOP-DOWN AI)
弱人工智能观点认为不可能制造出能真正地推理（REASONING）和解决问题（PROBLEM_SOLVING）的智能机器，<br>
这些机器只不过看起来像是智能的，但是并不真正拥有智能，也不会有自主意识。<br>
主流科研集中在弱人工智能上，并且一般认为这一研究领域已经取得可观的成就。强人工智能的研究则处于停滞不前的状态下。

>并不带有“弱”和“强”这两个汉字本身特有的贬义和褒义情感。`BOTTOM-UP`, `TOP-DOWN`

### 以iphone为标志的智能时代
计算器能够自动根据用户输入的任何数字计算出结果，为什么现在计算器不是人工智能？

另外iphone开启了智能手机时代，智能手机智能在哪里？


一般我们对物体的定义来自功能（类似动态语言的duck typing），如冰箱、电视、洗衣机、手表等，当这些司空见惯的家用电器加入了狂拽酷炫吊炸天的一些功能时就变成了如智能冰箱、智能电视、智能洗衣机、智能手表等，我们在说“智能”时，其实是在说“老铁，还有这种操作？”。有一天你的洗衣机坏了，请了一位比较健谈的师傅来修理后你知道了它的工作原理，虽然技术细节不知道，你也不会做，但是知道了有那么个东西负责干这个，潜移默化的就不觉得这玩意儿智能并且越来越少的在洗衣机前面加上“智能”了。

这也是为什么现在大家都不觉得使用的手机智能，因为你们知道了`加速计`、`陀螺仪`、`磁力计`、`GPS`、`气压计`、`距离传感器`、`光度传感器`等，例如通过GPS、磁力计、加速计等硬件收集的数据加上规划算法和前端UI就可以实现一个地图APP。因为可以通过你的实时位置作出规划，虽然没有人脸识别指纹解锁，但是也属于智能应用，搭载这个应用的设备也就是手机当然属于智能设配，只不过我们对这类APP已经熟练运用不觉得它“智能”而已。

能够落地的算法最终都会整合到框架中，而且一般都很快，所以不要地毯式盲目的学习算法，公司的产品都是使用框架实现的，小公司直接用，科技类大公司会对成熟框架进行解剖，然后加点针对自己公司业务的东西变成自己的框架。从框架里把算法翻出来看看python实现，一般框架都要求对关键部分编写完善的文档，基本上看完在自己实现一遍就会了，最坏的情况也能拿到一个算法名称，上算法网站有更丰富的教程。
前期关键是先要会使用框架写出功能，后期有时间你在公司也稳定了并且有兴趣，公司会给你时间去研究算法和论文，当然需要有产出。

现在学习人工智能没有用了吗？
有用。两个方面
1. AI企业要落地，对于AI企业而言，对业务的探索和用恰当的技术去解决实际问题才是首当其冲的。
2. 厉兵粟马，迎接AI。

人工智能行业在国内还远未达到十分乐观的地步，因此对于选择人工智能行业的人来说更加重要的是在工作岗位到达舒适区后对相关技术的敏感度和洞察力；<br>
具体来说，例如：
  1. 在你所处的人工智能领域智能感知或者无人驾驶，是否有最新论文解决公司当前问题？
  2. 如果有，你是否能够复现？
  3. 如果可以复现，是否可以使用真实数据将解决方案映射到公司的实际问题上？
