The way move from `discrete` to `continous` is to simply change the `active` function from `step` function to `sigmoid` function.

```python
import matplotlib.pyplot as plt
import numpy as np

def sigmoid(x: np.arange):
    return([1/(1+np.exp(-i)) for i in x])
    
x = np.arange(-10., 10., 0.2)
sig = sigmoid(x)
fig, axs = plt.subplots(1, 2, figsize=(10, 3))
axs[0].step([-10, 0, 10], [0, 0, 1])
axs[0].set_title('step function')
plt.plot(x, sig)
plt.title('sigmoid function')
plt.show()
```

cross entropy really says the following:<br>
If I have a bunch of events and a bunch of probabilities.<br>

If it's very likely, then we have a small cross entropy.<br><br>
